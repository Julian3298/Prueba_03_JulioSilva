```python
from itertools import cycle
personas_registradas = []
Certificados = {
    "Nacimiento" : "El Registro Civil certifica que el usuario {nombre} {apellido} \nidentificado con dni {dni} nacio en {pais_nacimiento} y tiene {edad} años",
    "Matrimonio" : "El Registro Civil certifica que el usuario {nombre} {apellido} \nidentificado con dni {dni} esta casado con {conyuge} \nel dia {fecha_matrimonio} y tiene {hijos} hijos",
}

def digito_verificador(dni):
  reversed_digits = map(int, reversed(str(dni)))
  factors = cycle(range(2, 8))
  s = sum(d * f for d, f in zip(reversed_digits, factors))
  return (-s) % 11

def validar_rut(dni):
    rut = str(dni) + "-" + str(digito_verificador(dni))
    return dni


def grabar_persona():
  dni = input("Ingrese el DNI sin digito verificador: ")
  digito= int(input("Ingrese digito verificador "))
  if digito_verificador(dni)==digito:
    nombre = input("Ingrese nombre: ")
    apellido = input("Ingrese apellido: ")
    edad = int(input("Ingrese la edad: "))
    pais_nacimiento = input("Ingrese el país de nacimiento: ")
    ciudad_nacimiento = input("Ingrese la ciudad de nacimiento: ")
    if edad >= 18:
            estado_civil = input("Ingrese su estado civil: ")
    else:
            estado_civil = None
    persona = [dni, nombre, apellido, edad, pais_nacimiento, ciudad_nacimiento, estado_civil]
    personas_registradas.append(persona)
    print("Se ha registrado correctamente.")
  else:
    print("Datos incorrectos")

def buscar_persona():
  dni = input("Ingrese el DNI de la persona que desea buscar: ")
  for persona in personas_registradas:
    if persona[0] == dni:
      print("****INFORMACIÓN****")
      print("DNI:", persona[0])
      print("Nombre: ", persona[1])
      print("Apellido: ", persona[2])
      print("Edad: ", persona[3])
      print("Pais de Nacimiento: ", persona[4])
      print("Ciudad de nacimiento: ", persona[5])
      print("Estado Civil: ", persona[6])

def imprimir_certificado():
  dni = input("Ingrese su dni sin digito verificador: ")
  for persona in personas_registradas:
    if persona[0] == dni:
      print("------CERTIFICADOS------")
      print("1.Nacimiento:")
      print("2.Matrimonio: ")
      certificado = int(input("Que tipo de certificado desea solicitar: "))
      if certificado == 1:
        if persona[4] == "argentina":
          print("---CERTIFICADO DE NACIMIENTO---")
          print("DNI:", persona[0])
          print("Nombre: ", persona[1])
          print("Apellido: ", persona[2])
          print("Edad: ", persona[3])
          print("Pais de Nacimiento: ", persona[4])
          print("Ciudad de nacimiento: ", persona[5])
          print("Estado Civil: ", persona[6])
          print("\n\t\t----------CERTIFICADO EMITIDO CON EXITO----------\n")
        else:
          print("El usuario no nacio en Argentina, no se puede emitir el certificado")
      elif certificado == 2:
        if persona[6] == "casado":
          conyuge = input("Ingrese el nombre de su conyuge: ")
          apellido_conyuge = input("Ingrese el apellido de su conyuge: ")
          conyuge = conyuge + " " + apellido_conyuge
          edad_conyuge= int(input("Ingrese edad de la conyuge"))
          fecha_matrimonio = input("Ingrese la fecha de matrimonio con el formato dd/mm/aaaa: ")
          hijos = int(input("Ingrese la cantidad de hijos que tiene: "))
          print("---CERTIFICADO DE MATRIMONIO---")
          print("DNI:", persona[0])
          print("Nombre: ", persona[1])
          print("Apellido: ", persona[2])
          print("Edad: ", persona[3])
          print("Pais de Nacimiento: ", persona[4])
          print("CASADO CON:")
          print("Nombre de la conyuge: ",conyuge)
          print("Edad: ",edad_conyuge)
          print("Fecha de matrimonio: ",fecha_matrimonio )
          print("Cantidad de Hijos: ", hijos)
          print("\n\t\t----------CERTIFICADO EMITIDO CON EXITO----------\n")
        else:
          print("El usuario no esta casado, no se puede emitir el certificado")
    else:
        print("El dni ingresado no se encuentra registrado")

def eliminar_persona():
  dni = input("Ingresa el dni de la persona que deseas eliminar: ")
  for persona in personas_registradas:
    if persona[0] == dni:
      personas_registradas.remove(persona)
      print("Ha sido eliminado exitosamente")
      return
    else:
      print("Esa persona no se encuentra registrada")

def salir():
  print("Gracias por usar el programa. ¡Hasta luego!")

def menu():
  while True:
        print("----- MENÚ -----")
        print("1. Grabar")
        print("2. Buscar")
        print("3. Imprimir certificados")
        print("4. Eliminar")
        print("5. Salir")
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
          grabar_persona()
        elif opcion == "2":
          buscar_persona()
        elif opcion == "3":
          imprimir_certificado()
        elif opcion == "4":
          eliminar_persona()
        elif opcion == "5":
          salir()
          break
        else:
            print("Opción inválida. Intente nuevamente.")

menu()
```
